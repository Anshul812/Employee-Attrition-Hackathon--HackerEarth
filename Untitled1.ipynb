```python
cat_col = ['Gender' , 'Relationship_Status' , 'Hometown' , 'Unit' , 'Decision_skill_possess' , 'Compensation_and_Benefits']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for i in cat_col:
    train_df[i] = le.fit_transform(train_df[i])
```

```python
train_df.head()
```

```python
train_df.info()
```

```python
train_df['Age'] = train_df.Age.fillna(train_df.Age.mean())
```

```python
train_df['Time_of_service'] = train_df.Time_of_service.fillna(train_df.Time_of_service.mean())
```

```python
train_df.Pay_Scale = train_df.Pay_Scale.fillna(train_df.Pay_Scale.mean())
```

```python
train_df.Work_Life_balance = train_df.Work_Life_balance.fillna(train_df.Work_Life_balance.mean())
```

```python
train_df.VAR2 = train_df.VAR2.fillna(train_df.VAR2.mean())
```

```python
train_df.VAR4 = train_df.VAR4.fillna(train_df.VAR4.mean())
```

```python
train_df.info()
```

```python
train_df.head()
```

```python
new_col = ['Gender', 'Age', 'Education_Level', 'Relationship_Status', 'Hometown',
       'Unit', 'Decision_skill_possess', 'Time_of_service',
       'Time_since_promotion', 'growth_rate', 'Travel_Rate', 'Post_Level',
       'Pay_Scale', 'Compensation_and_Benefits', 'Work_Life_balance', 'VAR1',
       'VAR2', 'VAR3', 'VAR4', 'VAR5', 'VAR6', 'VAR7']
```

```python
len(new_col)
```

```python
train_out = train_df.Attrition_rate
```

```python
train_df = train_df.drop('Attrition_rate', axis=1)
```

```python
mean_train = []

for i in new_col:
    mean_train.append(train_df[i].mean())
```

```python
std_train = []

for i in new_col:
    std_train.append(train_df[i].std())
```

```python
n=0
for i in new_col:
    train_df[i] = (train_df[i] - mean_train[n]) / std_train[n]
    n += 1
```

```python
train_df = train_df.drop('Employee_ID',axis=1)
```

```python
test_df = pd.read_csv("./Dataset/Test.csv")
```

```python
test_df.head()
```

```python
e_id = test_df.Employee_ID
```

```python
test_df = test_df.drop('Employee_ID' , axis=1)
```

```python
test_df.shape
```

```python
for i in cat_col:
    test_df[i] = le.fit_transform(test_df[i])
```

```python
test_df.head()
```

```python
test_df.info()
```

```python
test_df['Age'] = test_df.Age.fillna(test_df.Age.mean())

test_df['Time_of_service'] = test_df.Time_of_service.fillna(test_df.Time_of_service.mean())

test_df.Pay_Scale = test_df.Pay_Scale.fillna(test_df.Pay_Scale.mean())

test_df.Work_Life_balance = test_df.Work_Life_balance.fillna(test_df.Work_Life_balance.mean())

test_df.VAR2 = test_df.VAR2.fillna(test_df.VAR2.mean())

test_df.VAR4 = test_df.VAR4.fillna(test_df.VAR4.mean())
```

```python
test_df.head()
```

```python
test_df.info()
```

```python
test_df.head()
```

```python
n = 0
for i in new_col:
    test_df[i] = ( test_df[i] - mean_train[n] ) / std_train[n]
    n += 1
```

```python
test_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data , test_data , train_label , test_label = train_test_split(train_df , train_out , test_size=0.2 , random_state=42)
```

```python
print(train_data.shape , test_data.shape , train_label.shape , test_data.shape , test_df.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
model = LinearRegression()

model.fit(train_data , train_label)
model_pred = model.predict(test_data)

print((1-np.sqrt(mean_squared_error(y_pred=model_pred , y_true=test_label)))*100)
```

```python
test_out = model.predict(test_df)
out_df = pd.DataFrame({'Employee_ID':e_id , 'Attrition_rate':test_out})
out_df.to_csv("out_file.csv",index=False)
```
