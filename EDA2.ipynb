```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("./Dataset/Train.csv")
test = pd.read_csv("./Dataset/Test.csv")
```

```python
#AGE

data = [train , test]
for d in data:
    mean = train.Age.mean()
    std = train.Age.std()
    
    age_count = d.Age.isnull().sum()
    rand_age = np.random.randint(mean-std , mean+std , age_count)
    
    age_col = d.Age.copy()
    age_col[np.isnan(age_col)] = rand_age
    
    d.Age = age_col
    d.Age = d.Age.astype(int)
```

```python
#Age Category

dataset = [train , test]
for data in dataset:
    data.loc[data.Age <= 25 , 'Age'] = 0
    data.loc[(data.Age > 25) & (data.Age <= 35) , 'Age'] = 1
    data.loc[(data.Age > 35) & (data.Age <= 45) , 'Age'] = 2
    data.loc[(data.Age > 45) & (data.Age <= 55) , 'Age'] = 3
    data.loc[data.Age > 55 , 'Age'] = 4
    data.Age = data.Age.astype(int)
```

```python
#Gender
gender_map = {'M':1 , 'F':0}
for data in dataset:
    data.Gender = data.Gender.map(gender_map)
    data.Gender = data.Gender.astype(int)
```

```python
#Relationship_status
relation_map = {'Married':1 , 'Single':0}
for data in dataset:
    data.Relationship_Status = data.Relationship_Status.map(relation_map)
    data.Relationship_Status = data.Relationship_Status.astype(int)
```

```python
#Hometown
home_map = {'Franklin':0, 'Springfield':1, 'Clinton':2, 'Lebanon':3, 'Washington':4}

for data in dataset:
    data.Hometown = data.Hometown.map(home_map)
    data.Hometown = data.Hometown.astype(int)
```

```python
#Unit 
unit_map = {'IT':0, 'Logistics':1, 'Quality':2, 'Human Resource Management':3,
       'Purchasing':4, 'Sales':5, 'Production':6, 'Operarions':7,
       'Accounting and Finance':8, 'Marketing':9, 'R&D':10, 'Security':11}

for data in dataset:
    data.Unit = data.Unit.map(unit_map)
    data.Unit = data.Unit.astype(int)
```

```python
#Decision Skill Possess
decision_map = {'Conceptual':0, 'Analytical':1, 'Behavioral':2, 'Directive':3}

for data in dataset:
    data.Decision_skill_possess = data.Decision_skill_possess.map(decision_map)
    data.Decision_skill_possess = data.Decision_skill_possess.astype(int)
```

```python
#Time of Service

for data in dataset:
    mean = train.Time_of_service.mean()
    std = train.Time_of_service.std()
    
    tos_count = data.Time_of_service.isnull().sum()
    rand_tos = np.random.randint(mean-std , mean+std , tos_count)
    
    tos_col = data.Time_of_service.copy()
    tos_col[np.isnan(tos_col)] = rand_tos
    
    data.Time_of_service = tos_col
    data.Time_of_service = data.Time_of_service.astype(int)
```

```python
#Time Category

for data in dataset:
    data.loc[data.Time_of_service <= 8 , 'Time_of_service'] = 0
    data.loc[(data.Time_of_service > 8) & (data.Time_of_service <= 16) , 'Time_of_service'] = 1
    data.loc[(data.Time_of_service > 16) & (data.Time_of_service <= 24) , 'Time_of_service'] = 2
    data.loc[(data.Time_of_service > 24) & (data.Time_of_service <= 32) , 'Time_of_service'] = 3
    data.loc[data.Time_of_service > 32 , 'Time_of_service']=4
    data.Time_of_service = data.Time_of_service.astype(int)
```

```python
#Pay Scale
train.Pay_Scale = train.Pay_Scale.fillna(6)
test.Pay_Scale = test.Pay_Scale.fillna(6)
```

```python
#Pay SCale Category

for data in dataset:
    data.loc[data.Pay_Scale <= 3 , 'Pay_Scale']=0
    data.loc[(data.Pay_Scale > 3) & (data.Pay_Scale <= 7) , 'Pay_Scale']=1
    data.loc[(data.Pay_Scale > 7) & (data.Pay_Scale <= 10) , 'Pay_Scale']=2
```

```python
for data in dataset:
    data.Pay_Scale = data.Pay_Scale.astype(int)
```

```python
#Growth Rate

for data in dataset:
    data.loc[data.growth_rate <= 29 , 'growth_rate']=0
    data.loc[(data.growth_rate > 29) & (data.growth_rate <= 39) , 'growth_rate']=1
    data.loc[(data.growth_rate > 39) & (data.growth_rate <= 49) , 'growth_rate']=2
    data.loc[(data.growth_rate > 49) & (data.growth_rate <= 59) , 'growth_rate']=3
    data.loc[data.growth_rate > 59 , 'growth_rate']=4
```

```python
for data in dataset:
    data.growth_rate = data.growth_rate.astype(int)
```

```python
#Compensation and Benefits

compen_map = {'type2':0, 'type3':1, 'type0':2, 'type4':3, 'type1':4}
for data in dataset:
    data.Compensation_and_Benefits = data.Compensation_and_Benefits.map(compen_map)
    data.Compensation_and_Benefits = data.Compensation_and_Benefits.astype(int)
```

```python
#Work Life Balance

for data in dataset:
    data.Work_Life_balance = data.Work_Life_balance.fillna(2)
    data.Work_Life_balance = data.Work_Life_balance.astype(int)
```

```python
#VAR2
for data in dataset:
    data.VAR2 = data.VAR2.fillna(0.7516)
```

```python
#VAR2 , VAR3
for data in dataset:
    data.VAR2 = round(data.VAR2)
    data.VAR3 = round(data.VAR3)
    data.VAR2 = data.VAR2.astype(int)
    data.VAR3 = data.VAR3.astype(int)
```

```python
for data in dataset:
    data.VAR4 = data.VAR4.fillna(2)
    data.VAR4 = data.VAR4.astype(int)
```

```python
from sklearn.model_selection import train_test_split
train_data , test_data , train_label , test_label = train_test_split(train , train_out , test_size=0.2 , random_state=42)
```

```python
print(train_data.shape , test_data.shape , train_label.shape , test_label.shape)
```

```python
#Lets build ML model
import tensorflow  as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Model
```

```python
inp = layers.Input(shape=(22,))

x = layers.Dense(11 , activation='relu')(inp)
x = layers.Dense(5 , activation='relu')(x)

out = layers.Dense(1 , activation='relu')(x)

model = Model(inp ,out)
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error' , optimizer='sgd')
```

```python
model.fit(train_data , train_label , epochs=5 , verbose=1 , validation_data=(test_data , test_label))
```

```python
test_out = model.predict(test)
test_out = test_out.reshape(test_out.shape[0])

out_df = pd.DataFrame({'Employee_ID':e_id , 'Attrition_rate':test_out})
out_df.to_csv("out_file.csv",index=False)
```
